```python
df = pd.read_csv("MedQuAD-master.csv",encoding="ISO-8859-1")
```

```python
answers = df['Answer'].to_numpy()
answers
```

```python
# Save the answers embedding
import pickle
def save_docs_emb():
  with open('doc_emb.pkl', 'wb') as f:
      pickle.dump(doc_emb, f)

def load_docs_emb():
  with open('doc_emb.pkl', 'rb') as f:
    return pickle.load(f)
```

```python
docs = answers

doc_emb = model.encode(docs)
save_docs_emb()
# doc_emb = load_docs_emb()
```

```python
query = "What is corona virus"
query_emb = model.encode(query)
scores = util.dot_score(query_emb, doc_emb)[0].cpu().tolist()

idx = np.argmax(scores)
tmp_ans = answers[idx].split('\n')
tmp_ans_emb = model.encode(tmp_ans)
tmp_score = util.dot_score(query_emb, tmp_ans_emb)[0].cpu().tolist()

print(tmp_ans[np.argmax(tmp_score)])
```
